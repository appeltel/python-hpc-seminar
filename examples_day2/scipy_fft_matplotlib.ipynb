## SciPy FFT and matplotlib example

This example is derived from the [SciPy documentation](https://docs.scipy.org/doc/scipy-1.0.0/reference/tutorial/fftpack.html) and demonstrates how to run a fast Fourier transform on a small set of data and plot the results with [matplotlib](https://matplotlib.org/)

First import numpy and scipy fft:

```python
import numpy as np
from scipy.fftpack import fft
```

Now import matplotlib. Note the jupyter notebook directive `%matplotlib inline` which is required to display the plot inline as part of the notebook.

```python
%matplotlib inline

import matplotlib
import matplotlib.pyplot as plt
```

Set the number of sample points (N) and sample spacing (T)

```python
N = 6000
T = 1.0 / 1600.0
```

Now generate an array of data from a handful of actual signals (sine waves) and some generated noise from a random distribution.

```python
x = np.linspace(0.0, N*T, N)
y = (
    20 * np.sin(50.0 * 2.0*np.pi*x) +
    10 * np.sin(80.0 * 2.0*np.pi*x) +
    40 * np.sin(315.0 * 2.0*np.pi*x) +
    np.random.uniform(-50., 50., size=len(x))
)
```

Let's look at the data itself first:

```python
plt.figure(figsize=(15,8))
plt.scatter(x, y)
plt.grid()
plt.show()
```

Perform the FFT of this dataset

```python
yf = fft(y)
xf = np.linspace(0.0, 1.0/(2.0*T), N//2)
```

...and plot the results

```python
plt.figure(figsize=(15,8))
plt.plot(xf, 2.0/N * np.abs(yf[0:N//2]))
plt.grid()
plt.show()
```
